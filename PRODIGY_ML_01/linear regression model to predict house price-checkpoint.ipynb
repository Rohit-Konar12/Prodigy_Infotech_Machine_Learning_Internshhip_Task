```python
import pandas as pd
```

```python
data = pd.read_csv(r"D:\BackupOneDrive\Desktop\JN\PRODIGY_ML_01\train.csv")
```

```python
data.head()
```

```python
print(data.isnull().sum())
```

```python
data = data.drop(['Alley', 'PoolQC', 'Fence', 'MiscFeature'], axis=1)
```

```python
lotfrontage_median = data['LotFrontage'].median()
data['LotFrontage'] = data['LotFrontage'].fillna(lotfrontage_median)
```

```python
mode_value = data['MasVnrType'].mode()[0]
data['MasVnrType'] = data['MasVnrType'].fillna(mode_value)
```

```python
print(data.isnull().sum().head())
```

```python
data['TotalBath'] = data['FullBath'] + 0.5 * data['HalfBath']
features = data[['GrLivArea', 'BedroomAbvGr', 'TotalBath']]
target = data['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, predictions)
r2 = r2_score(y_test, predictions)
print(f"Mean Squared Error: {mse}")
print(f"R-Squared: {r2}")
```

```python
import matplotlib.pyplot as plt

res = y_test - predictions
plt.scatter(model.predict(X_test), res, alpha=0.5)
plt.axhline(0, linestyle='--')
plt.xlabel('Predicted price')
plt.ylabel('Residuals')
plt.title('Residuals vs Predicted')
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np

predicted_prices = np.array([210000, 240000, 265000, 295000, 325000, 355000, 390000])
residuals = np.array([-10000, 10000, 5000, 5000, -5000, 5000, 10000])  # actual - predicted

fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(predicted_prices.min() - 20000, predicted_prices.max() + 20000)
ax.set_ylim(residuals.min() - 10000, residuals.max() + 10000)
ax.set_xlabel('Predicted Prices')
ax.set_ylabel('Residuals (Actual - Predicted)')
ax.set_title('Residual Plot of Predicted House Prices')
scat = ax.scatter([], [], color='green', label='Residual Points')
ax.axhline(0, color='red', linestyle='--', label='Zero Residual Line')
ax.legend()
ax.grid(True)

def init():
    scat.set_offsets(np.empty((0, 2)))  # Empty 2D array
    return scat,

def animate(i):
    data = np.column_stack((predicted_prices[:i+1], residuals[:i+1]))
    scat.set_offsets(data)
    return scat,

ani = animation.FuncAnimation(fig, animate, init_func=init,
                              frames=len(predicted_prices), interval=800, blit=True)

ani.save('residual_vs_predicted_price.gif', writer='pillow', fps=1)
```
